Run that cell and wait for it to finish.

Next we'll try using Plots to draw a graph, but first we need some data. The code below sets up a series of 20 random numbers in the variable <i>y</i>, along with an index from 1 to 20 in the variable <i>x</i>. We then call <i>plot</i> with the variables <i>x</i> and <i>y</i> and a label set to the text "my label".

Running this cell will give you a line graph similar to the one below.

```julia
x = 1:20; y = rand(20); # These are the plotting data
plot(x,y, label="my label")
```

```julia
savefig("simpleplot.png")
```

## Using real data
We can see just how easy it is to create

```julia
using CSV
using HTTP
```

```julia

file = HTTP.get("https://coded2.herokuapp.com/datavizpandas/london2018.csv")
```

```julia

d = CSV.read(IOBuffer(file.body))
```

```julia
bar(d.Month,d.Tmax)
```

```julia
savefig("bartmax.png")
```

```julia
plot(d.Month, d.Tmax)
```

```julia
savefig("linetmax.png")
```

```julia
scatter(d.Tmax, d.Sun)
```

```julia
savefig("scattertmax.png")
```

```julia
x = 1:5; y = rand(5); # These are the plotting data
pie(x,y, label="123")
```

```julia
savefig("pierand.png")
```

```julia
using DataFrames
file = HTTP.get("https://coded2.herokuapp.com/datavizpandas/londonweather.csv")
d2 = CSV.read(IOBuffer(file.body))
d3 = groupby(d2,:Month)
d3[6]
```

```julia
histogram(d3[6].Tmax, bins=10, label="Maximum July Temperature")
```

```julia
savefig("histjuly.png")
```

```julia
scatter(d3[6].Tmax, d3[6].Sun, label="Sun and Max Temperature")
```

```julia
savefig("scattersun.png")
```
